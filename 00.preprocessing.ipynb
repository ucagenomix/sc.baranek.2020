# wt
[Back to top](#wt)

```python
wt = sc.read_10x_mtx("/data/10x_data/10x_paget/wt/filtered_gene_bc_matrices/mm10", var_names='gene_symbols', cache=True) 
wt.var_names_make_unique()  # this is unnecessary if using 'gene_ids'
wt.obs['sample'] = 'wt'
wt
```

```python
sc.pl.highest_expr_genes(wt, n_top=10)
```

```python
sc.pp.filter_cells(wt, min_genes=0)
sc.pp.filter_genes(wt, min_cells=3)

mito_genes = wt.var_names.str.startswith('mt-')
# for each cell compute fraction of counts in mito genes vs. all genes
# the `.A1` is only necessary as X is sparse (to transform to a dense array after summing)
wt.obs['percent_mito'] = np.sum(
    wt[:, mito_genes].X, axis=1).A1 / np.sum(wt.X, axis=1).A1

ribo_genes = wt.var_names.str.startswith('Rp')
# for each cell compute fraction of counts in mito genes vs. all genes
# the `.A1` is only necessary as X is sparse (to transform to a dense array after summing)
wt.obs['percent_ribo'] = np.sum(
    wt[:, ribo_genes].X, axis=1).A1 / np.sum(wt.X, axis=1).A1

# add the total counts per cell as observations-annotation to adata
wt.obs['n_counts'] = wt.X.sum(axis=1).A1

sc.pl.violin(wt, ['n_genes', 'n_counts', 'percent_mito', 'percent_ribo'],
             jitter=0.4, multi_panel=True)
```

```python
wt = wt[wt.obs['percent_mito'] < 0.05, :]
```

```python
wt.shape
```

```python
# Remove unwanted genes (ribosoaml and mitochondrial genes)
genes_names = wt.var.index.tolist()
keep_genes = [i for i in genes_names if not i.startswith('mt-')]
wt = wt[:,keep_genes]
genes_names = wt.var.index.tolist()
keep_genes = [i for i in genes_names if not i.startswith('Rps')]
wt = wt[:,keep_genes]
genes_names = wt.var.index.tolist()
keep_genes = [i for i in genes_names if not i.startswith('Rpl')]
wt = wt[:,keep_genes]
wt.shape
```

```python
# scrublet
scrub = scr.Scrublet(wt.X, expected_doublet_rate=0.05) 
doublet_scores, predicted_doublets = scrub.scrub_doublets()
wt.obs['doublet_scores'] = doublet_scores
wt.obs['predicted_doublets'] = predicted_doublets
scrub.plot_histogram();
```

```python
# doubletDetection
clf = doubletdetection.BoostClassifier()
labels = clf.fit(wt.X).predict()
wt.obs['doubletDetection'] = labels
```

```python
sc.pp.normalize_per_cell(wt, counts_per_cell_after=1e4) # Normalize all data
sc.pp.log1p(wt) # log transform the data
wt.raw = wt

sc.pl.scatter(wt, x='n_counts', y='percent_mito')
sc.pl.scatter(wt, x='n_counts', y='n_genes')
```
